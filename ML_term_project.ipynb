```python
from sklearn.decomposition import SparsePCA # importing the sprace principal component analysis library
from sklearn.datasets import load_digits # import the load digits dataset
import numpy as np #importing numpy
```

```python
digits = load_digits() #storing the dataset as an array
print(digits.data.shape) # printing the shape of the dataset

digits.data # displaying the data
```

```python
sparse_pca = SparsePCA(n_components=6, alpha = 0.1)  # applying sparse pca on the data set and setting the number of components to any arbitrary number 
sparse_pca.fit_transform(digits.data/255) # scaling the data by finding mean and variance
print("Shape of data: ", sparse_pca.components_.shape) # printing the shape of the dimension reduced data
print("Components: ", sparse_pca.components_) # printing the components of the data
```

```python
components = np.asarray(sparse_pca.components_) # storing the components found by spca in an array
print("Variances of components: ", np.var(components, axis = 1)) #using the var function to find the variance of each component
```

```python
from sklearn.decomposition import PCA # importing the Principal component analysis library
```

```python
pca = PCA(n_components=6) # doing pca on the same dataset
pca.fit_transform(digits.data/255) # scaling the data in a similar manner as done for spca
print("Shape of data: ", pca.components_.shape) #shape of the data after dimensionality reduction
print("Compoents: ", pca.components_) #printing the components after pca
```

```python
pca_components = np.asarray(pca.components_) # storing the components of data after applying pca
print("Variances of components " ,np.var(pca_components, axis = 1))  # finding variance of each component of the data
```
